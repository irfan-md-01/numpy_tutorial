```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.bar(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
x=['a','b','c','d','e']
y=[2,4,6,8,10]
plt.pie(x,y)
plt.show()
```
